```python
query = '''WITH MATCH_DAY AS(
	SELECT RTRIM(LTRIM(S_ACNO)) AS S_ACNO,
			MTH_DATE,
			FEE
	FROM ODS.dbo.ST_MATCH_D),
CUSTOMER_INFO AS(
	SELECT RTRIM(LTRIM(S_IDNO)) AS S_IDNO,
			RTRIM(LTRIM(S_ACNO)) AS S_ACNO,
			S_NAME,
			S_BIRTH,
			RADDR_O,
			CADDR_O,
			CASE WHEN S_IDNO LIKE '[A-Z]%' AND SUBSTRING(S_IDNO, 2, 1) = '1' THEN 'M'
					WHEN S_IDNO LIKE '[A-Z]%' AND SUBSTRING(S_IDNO, 2, 1) = '2' THEN 'F' ELSE 'N' END AS GENDER,
			CASE WHEN ISDATE(S_BIRTH) = 1 THEN DATEDIFF(YEAR, S_BIRTH, CONVERT(DATE, GETDATE())) ELSE NULL END AS [AGE]
	FROM ODS.dbo.ST_CUSTOMER
	WHERE S_IDNO LIKE '[A-Z]%' AND SUBSTRING(S_IDNO, 2, 1) IN ('1', '2') AND ISNUMERIC(RIGHT(S_IDNO, 8)) = 1)
SELECT C.S_IDNO,
		C.GENDER,
		C.AGE,
		C.S_NAME,
		LEFT(C.DATE_YMD, 6) AS YYYY_MM,
		C.ADDRESS_BORN,
		C.ADDRESS_COMM,
		C.LABEL_SLEEP,
		COUNT(*) AS TOTAL,
		SUM(C.FEE) AS FEE 
FROM(
SELECT MATCH_DAY.S_ACNO AS S_ACNO,
		MATCH_DAY.MTH_DATE AS DATE_YMD,
		MATCH_DAY.FEE AS FEE,
		CUSTOMER_INFO.S_IDNO AS S_IDNO,
		CUSTOMER_INFO.S_NAME AS S_NAME,
		CUSTOMER_INFO.GENDER AS GENDER,
		CUSTOMER_INFO.AGE AS AGE,
		CUSTOMER_INFO.RADDR_O AS ADDRESS_BORN,
		CUSTOMER_INFO.CADDR_O AS ADDRESS_COMM,
		CASE WHEN LAST_VALUE(MATCH_DAY.MTH_DATE) OVER (PARTITION BY S_IDNO ORDER BY MATCH_DAY.MTH_DATE) <= '20190102' THEN 1 ELSE 0 END AS LABEL_SLEEP
FROM MATCH_DAY
INNER JOIN CUSTOMER_INFO ON MATCH_DAY.S_ACNO = CUSTOMER_INFO.S_ACNO)C
GROUP BY C.S_IDNO, LEFT(C.DATE_YMD, 6), C.GENDER, C.AGE, C.ADDRESS_BORN, C.ADDRESS_COMM, C.S_NAME, C.LABEL_SLEEP'''
```
